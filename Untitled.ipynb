```python
import re

text = "Hello world"
pattern = r"Hello"
pattern2 = r"d$"

if re.match(pattern, text):
    print("Matched!")
else:
    print("Not matched.")

if re.search(pattern2, text):
    print("Matched!")
else:
    print("Not matched.")
```

Symbol	Meaning	Example
- .	----------->    Any character except newline                ------------------------------>	a.c → matches "abc", "a1c"
- \d	----------->    Digits (0–9)	                        ------------------------------>	\d\d → matches "42"
- \w	----------->    Letters, digits, underscore	            ------------------------------>  \w\w\w  matches "cat"
- \s	----------->    Whitespace (space, tab, newline)	    ------------------------------>    \s → matches " "
- +	----------->    One or more	\d+                             ------------------------------> "12345"
- *	----------->    Zero or more	                            ------------------------------> \w* → "hello" or ""
- {n}	----------->    Exactly n repetitions                   ------------------------------>	\d{4} → "2024"
- ^	----------->    Start of string                             ------------------------------>	^Hello
- $	----------->    End of string                               ------------------------------>	world$

| Function       | Purpose                       |
| -------------- | ----------------------------- |
| `re.match()`   | Checks beginning only         |
| `re.search()`  | Finds first match anywhere    |
| `re.findall()` | Returns ALL matches in a list |
| `re.sub()`     | Replace matches               |

**. means "any character"**

**\. means "a literal dot."**

| Pattern | Means                   |
| ------- | ----------------------- |
| `\.`    | match a literal dot `.` |
| `\(`    | match literal `(`       |
| `\+`    | match literal `+`       |

| Pattern | Means                                  |
| ------- | -------------------------------------- |
| `\d`    | digit (0-9)                            |
| `\w`    | word character (letters, numbers, `_`) |
| `\s`    | whitespace (space, tab, newline)       |
| `\D`    | NOT digit                              |
| `\W`    | NOT word character                     |
| `\S`    | NOT whitespace                         |


| Pattern       | Matches                 |
| ------------- | ----------------------- |
| `[abc]`       | matches **a OR b OR c** |
| `[0-9]`       | digits                  |
| `[A-Z]`       | uppercase letters       |
| `[A-Za-z0-9]` | any letter or digit     |


- **Important rule: everything inside [ ] is taken literally**
- **does NOT mean “word char + plus sign”.**
- **It means: one character that could be: w, +, etc.**
- **Because inside [], + loses its meaning**

| Symbol  | Meaning                | Example                               |
| ------- | ---------------------- | ------------------------------------- |
| `+`     | one or more            | `\d+` → `12345`                       |
| `*`     | zero or more           | `\d*` → `""` or `"123"`               |
| `?`     | zero or one (optional) | `colou?r` → `"color"` or `"colour"`   |
| `{n}`   | exactly n times        | `\d{4}` → `"2024"`                    |
| `{n,m}` | between n and m times  | `\d{2,4}` → `"12"`, `"123"`, `"1234"` |


```python
class EquilateralTriangle:
    def __init__(self, lenght: float):
        self.lenght = float(lenght)
        
    def perimeter(self):
        return 3 * (self.lenght)
    
    def area(self):
        return (3 ** 0.5 / 4) * (self.lenght) ** 2
```

```python
triangle1 = EquilateralTriangle(3)
print(f"The Area of a Equilateral Triangle of a Side Lentgh of {triangle1.lenght} is: {triangle1.area()} m2")
print(f"The Perimeter of a Equilateral Triangle of a Side Lentgh of {triangle1.lenght} is: {triangle1.perimeter()} m")
```
